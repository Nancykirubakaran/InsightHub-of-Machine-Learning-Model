# Text Classification 

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.feature_extraction.text import TfidfVectorizer # Term Frequency 
from sklearn.metrics import classification_report
```

```python
data = pd.read_csv("Test.csv")
```

```python
data
```

```python
data_text = data['text']
data_targ = data['label']
```

```python
vect  = TfidfVectorizer()
```

```python
train_data = vect.fit_transform(data_text)
```

```python
len(vect.get_feature_names_out())
```

```python
len(train_data.toarray()[0])
```

```python
model = SVC(kernel="linear")
```

```python
model.fit(train_data,data_targ)
```

```python
SVC(kernel='linear')
```

```python
predicted_data = model.predict(train_data)
```

```python
print(classification_report(data_targ, predicted_data))
```
